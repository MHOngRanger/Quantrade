# Strategy 1: Pairs Trading (Statistical Arbitrage)

**Idea**: Find two cointegrated stocks. When their price spread deviates beyond a threshold,
short the expensive one and long the cheap one, betting the spread reverts to mean.

**Logic**:
1. Test for cointegration using Engle-Granger test
2. Estimate hedge ratio via OLS regression
3. Compute spread = log(A) - β * log(B)
4. Z-score the spread (rolling)
5. Entry: |z| > 2, Exit: |z| < 0.5, Stop: |z| > 3.5

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller
from itertools import combinations
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
# --- Parameters ---
START = '2015-01-01'
END   = '2024-12-31'
ZSCORE_ENTRY  = 2.0
ZSCORE_EXIT   = 0.5
ZSCORE_STOP   = 3.5
LOOKBACK      = 60    # rolling window for z-score (trading days)
TRANSACTION_COST = 0.001  # 10 bps per leg

# Candidate pairs: same-sector stocks known to co-move
CANDIDATE_PAIRS = [
    ('KO', 'PEP'),    # Beverages
    ('GS', 'MS'),     # Investment banks
    ('XOM', 'CVX'),   # Energy majors
    ('MCD', 'YUM'),   # Fast food
    ('JPM', 'BAC'),   # Big banks
    ('AMZN', 'EBAY'), # E-commerce
    ('HD', 'LOW'),    # Home improvement
]
```

```python
# --- Download data ---
all_tickers = list({t for pair in CANDIDATE_PAIRS for t in pair})
raw = yf.download(all_tickers, start=START, end=END, auto_adjust=True, progress=False)
prices = raw['Close'].dropna()
log_prices = np.log(prices)
print(f"Downloaded {len(prices)} days for {len(prices.columns)} tickers")
prices.tail(3)
```

```python
# --- Cointegration screening ---
def test_cointegration(s1, s2, significance=0.05):
    """Engle-Granger cointegration test. Returns (p_value, hedge_ratio)."""
    score, pvalue, _ = coint(s1, s2)
    # OLS hedge ratio
    X = sm.add_constant(s2)
    model = sm.OLS(s1, X).fit()
    beta = model.params.iloc[1]
    return pvalue, beta

print("Cointegration Test Results (in-sample: 2015-2019)")
print("-" * 55)
results = {}
train_end = '2019-12-31'
for a, b in CANDIDATE_PAIRS:
    if a not in log_prices.columns or b not in log_prices.columns:
        continue
    s1 = log_prices.loc[:train_end, a]
    s2 = log_prices.loc[:train_end, b]
    pval, beta = test_cointegration(s1, s2)
    cointed = pval < 0.05
    results[(a, b)] = {'pvalue': pval, 'beta': beta, 'cointegrated': cointed}
    print(f"{a}/{b:5s}  p={pval:.4f}  beta={beta:.3f}  {'✓ COINTEGRATED' if cointed else '✗'}")
```

```python
# --- Backtest engine for a single pair ---
def backtest_pair(ticker_a, ticker_b, hedge_ratio, log_px,
                  lookback=LOOKBACK, entry=ZSCORE_ENTRY,
                  exit_z=ZSCORE_EXIT, stop=ZSCORE_STOP, tc=TRANSACTION_COST):
    """
    Vectorised backtest.
    spread = log(A) - beta * log(B)
    Position: +1 = long spread (long A, short B)
              -1 = short spread (short A, long B)
    """
    df = pd.DataFrame({
        'a': log_px[ticker_a],
        'b': log_px[ticker_b]
    }).dropna()

    df['spread'] = df['a'] - hedge_ratio * df['b']
    df['mean']   = df['spread'].rolling(lookback).mean()
    df['std']    = df['spread'].rolling(lookback).std()
    df['zscore'] = (df['spread'] - df['mean']) / df['std']

    # Signal generation (shift 1 to avoid look-ahead)
    pos = np.zeros(len(df))
    current = 0
    for i in range(lookback, len(df)):
        z = df['zscore'].iloc[i]
        if np.isnan(z):
            continue
        if current == 0:
            if z >  entry: current = -1   # short spread
            if z < -entry: current =  1   # long spread
        elif current == 1:
            if z >  stop:  current =  0   # stop
            if z <  exit_z: current = 0   # profit take
        elif current == -1:
            if z < -stop:  current =  0   # stop
            if z > -exit_z: current = 0   # profit take
        pos[i] = current

    df['pos'] = pos

    # P&L: log returns of spread × position
    df['spread_ret'] = df['spread'].diff()
    df['gross_pnl']  = df['pos'].shift(1) * df['spread_ret']

    # Transaction costs on position changes
    trades = df['pos'].diff().abs()
    df['tc_cost'] = trades * tc * 2  # two legs
    df['net_pnl'] = df['gross_pnl'] - df['tc_cost']
    df['equity']  = df['net_pnl'].cumsum()

    return df


def performance_stats(pnl_series, label='Strategy'):
    """Print key backtest stats."""
    daily = pnl_series.dropna()
    ann_ret  = daily.mean() * 252
    ann_vol  = daily.std() * np.sqrt(252)
    sharpe   = ann_ret / ann_vol if ann_vol != 0 else 0
    cum      = daily.cumsum()
    drawdown = cum - cum.cummax()
    max_dd   = drawdown.min()
    n_trades = (daily != 0).sum()
    print(f"\n{'='*45}")
    print(f"  {label}")
    print(f"{'='*45}")
    print(f"  Ann. Return : {ann_ret*100:.2f}%")
    print(f"  Ann. Vol    : {ann_vol*100:.2f}%")
    print(f"  Sharpe Ratio: {sharpe:.3f}")
    print(f"  Max Drawdown: {max_dd*100:.2f}%")
    print(f"  Active Days : {n_trades}")
    return {'ann_ret': ann_ret, 'ann_vol': ann_vol, 'sharpe': sharpe, 'max_dd': max_dd}
```

```python
# --- Run backtest on all cointegrated pairs (out-of-sample: 2020-2024) ---
oos_start = '2020-01-01'
fig, axes = plt.subplots(len(CANDIDATE_PAIRS), 2, figsize=(16, 3.5 * len(CANDIDATE_PAIRS)))
all_stats = []

for idx, (a, b) in enumerate(CANDIDATE_PAIRS):
    if a not in log_prices.columns or b not in log_prices.columns:
        continue
    info = results.get((a, b), {})
    beta = info.get('beta', 1.0)

    # Full period backtest (both in-sample and out-of-sample)
    res = backtest_pair(a, b, beta, log_prices)
    oos = res.loc[oos_start:]

    stats = performance_stats(oos['net_pnl'], f"{a}/{b} OOS 2020-2024")
    stats['pair'] = f"{a}/{b}"
    all_stats.append(stats)

    ax1, ax2 = axes[idx]
    ax1.plot(oos.index, oos['zscore'], color='steelblue', lw=0.8)
    ax1.axhline( ZSCORE_ENTRY, color='red',   ls='--', lw=0.8, label='Entry')
    ax1.axhline(-ZSCORE_ENTRY, color='red',   ls='--', lw=0.8)
    ax1.axhline( ZSCORE_EXIT,  color='green', ls='--', lw=0.8, label='Exit')
    ax1.axhline(-ZSCORE_EXIT,  color='green', ls='--', lw=0.8)
    ax1.set_title(f"{a}/{b}  |  Z-Score (OOS)", fontsize=11)
    ax1.legend(fontsize=8); ax1.set_ylabel('Z-Score')

    ax2.plot(oos.index, oos['equity'], color='darkorange', lw=1.2)
    ax2.fill_between(oos.index, oos['equity'], alpha=0.15, color='darkorange')
    ax2.axhline(0, color='grey', lw=0.6)
    ax2.set_title(f"{a}/{b}  |  Cumulative Log P&L", fontsize=11)
    ax2.set_ylabel('Cum P&L')

plt.tight_layout()
plt.savefig('../data/pairs_trading_results.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# --- Summary table ---
summary = pd.DataFrame(all_stats).set_index('pair')
summary['ann_ret'] = summary['ann_ret'].map('{:.2%}'.format)
summary['ann_vol'] = summary['ann_vol'].map('{:.2%}'.format)
summary['sharpe']  = summary['sharpe'].map('{:.3f}'.format)
summary['max_dd']  = summary['max_dd'].map('{:.2%}'.format)
summary.columns    = ['Ann. Return', 'Ann. Vol', 'Sharpe', 'Max DD']
print("\nOut-of-Sample Summary (2020-2024)")
print(summary.to_string())
```

```python
# --- Deep dive: best pair ---
best_pair_name = max(all_stats, key=lambda x: float(x['sharpe']))
a, b = best_pair_name['pair'].split('/')
beta = results[(a, b)]['beta']
res = backtest_pair(a, b, beta, log_prices)

fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)
axes[0].plot(res.index, np.exp(log_prices[a]) / np.exp(log_prices[a]).iloc[0], label=a)
axes[0].plot(res.index, np.exp(log_prices[b]) / np.exp(log_prices[b]).iloc[0], label=b)
axes[0].set_title(f"{a}/{b} — Normalised Prices"); axes[0].legend()

axes[1].plot(res.index, res['zscore'], color='steelblue', lw=0.8)
axes[1].axhline(ZSCORE_ENTRY, color='red', ls='--', lw=0.8)
axes[1].axhline(-ZSCORE_ENTRY, color='red', ls='--', lw=0.8)
axes[1].set_title('Z-Score of Spread')

axes[2].plot(res.index, res['equity'], color='darkorange', lw=1.2)
axes[2].fill_between(res.index, res['equity'], alpha=0.15, color='darkorange')
axes[2].set_title('Cumulative P&L')
plt.tight_layout()
plt.show()
performance_stats(res['net_pnl'], f"{a}/{b} FULL PERIOD")
```
