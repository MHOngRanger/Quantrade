# Strategy 2: Cross-Sectional Momentum

**Idea**: Stocks that outperformed in the past (12-1 months) tend to keep outperforming.
The '1' skip avoids the short-term reversal effect documented by Jegadeesh & Titman.

**Logic**:
1. Universe: ~100 liquid large-cap US stocks
2. Every month, compute 12-1 month total return for each stock
3. Rank stocks; go long top quintile, short bottom quintile
4. Equal-weight within each bucket, rebalance monthly
5. Net of transaction costs

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 5)
```

```python
# --- Parameters ---
START = '2010-01-01'
END   = '2024-12-31'
LOOKBACK_MONTHS  = 12   # total lookback
SKIP_MONTHS      = 1    # skip most recent month (reversal avoidance)
QUINTILE         = 0.2  # top/bottom 20%
TRANSACTION_COST = 0.001  # 10 bps per leg

# ~100 large-cap S&P 500 stocks across sectors
UNIVERSE = [
    # Tech
    'AAPL','MSFT','GOOGL','META','NVDA','AMZN','TSLA','ORCL','IBM','INTC',
    'QCOM','TXN','AMD','AMAT','MU',
    # Financials
    'JPM','BAC','GS','MS','WFC','C','BLK','AXP','USB','PNC',
    # Healthcare
    'JNJ','UNH','PFE','ABBV','MRK','LLY','BMY','AMGN','MDT','ABT',
    # Consumer Staples
    'KO','PEP','PG','WMT','COST','MCD','NKE','SBUX','YUM','CL',
    # Energy
    'XOM','CVX','COP','EOG','SLB','PSX','VLO','MPC','HES','OXY',
    # Industrials
    'BA','CAT','GE','HON','MMM','UPS','RTX','LMT','DE','EMR',
    # Materials & Utilities
    'LIN','APD','NEM','FCX','NEE','DUK','SO','D','EXC','AEP',
    # Benchmark
    'SPY'
]
```

```python
# --- Download monthly prices ---
raw = yf.download(UNIVERSE, start=START, end=END, auto_adjust=True,
                  progress=False, interval='1mo')
px_monthly = raw['Close'].dropna(how='all')
# Keep only tickers with enough history
px_monthly = px_monthly.loc[:, px_monthly.notna().sum() > 100]
monthly_ret = px_monthly.pct_change()
print(f"Universe: {len(px_monthly.columns)-1} stocks + SPY  |  {len(px_monthly)} months")
```

```python
# --- Build momentum signal ---
# Momentum = cumulative return from t-12 to t-1 (skip last month)
def momentum_signal(px, lookback=LOOKBACK_MONTHS, skip=SKIP_MONTHS):
    """12-1 month momentum: return from [t-lookback] to [t-skip]."""
    lag_long  = px.shift(skip)
    lag_short = px.shift(lookback)
    return lag_long / lag_short - 1

signal = momentum_signal(px_monthly)

# Drop SPY from stock universe for signal computation
stock_cols = [c for c in signal.columns if c != 'SPY']
stock_signal = signal[stock_cols]
```

```python
# --- Backtest: monthly rebalancing ---
port_returns = []
dates = []

for i in range(LOOKBACK_MONTHS + 1, len(px_monthly)):
    date = px_monthly.index[i]
    sig_row = stock_signal.iloc[i].dropna()
    if len(sig_row) < 20:
        continue

    ranked = sig_row.rank(pct=True)
    longs  = ranked[ranked >= 1 - QUINTILE].index.tolist()
    shorts = ranked[ranked <= QUINTILE].index.tolist()

    # Next month returns (forward-looking, execution at month-end close)
    fwd_ret = monthly_ret[stock_cols].iloc[i]

    long_ret  = fwd_ret[longs].mean()
    short_ret = fwd_ret[shorts].mean()

    gross_ret = (long_ret - short_ret) / 2  # dollar neutral
    tc_cost   = TRANSACTION_COST * 2        # rough turnover estimate
    net_ret   = gross_ret - tc_cost

    port_returns.append(net_ret)
    dates.append(date)

port_ret = pd.Series(port_returns, index=dates, name='Momentum')
spy_ret  = monthly_ret['SPY'].reindex(dates)

print(f"Backtest: {dates[0].date()} to {dates[-1].date()}")
print(f"Months: {len(dates)}")
```

```python
# --- Performance stats ---
def stats(ret, label=''):
    ann_ret = ret.mean() * 12
    ann_vol = ret.std() * np.sqrt(12)
    sharpe  = ann_ret / ann_vol if ann_vol != 0 else 0
    cum     = (1 + ret).cumprod()
    rolling_max = cum.cummax()
    max_dd  = ((cum - rolling_max) / rolling_max).min()
    calmar  = ann_ret / abs(max_dd) if max_dd != 0 else 0
    hit     = (ret > 0).mean()
    print(f"{'='*40}\n  {label}\n{'='*40}")
    print(f"  Ann. Return  : {ann_ret*100:.2f}%")
    print(f"  Ann. Vol     : {ann_vol*100:.2f}%")
    print(f"  Sharpe Ratio : {sharpe:.3f}")
    print(f"  Max Drawdown : {max_dd*100:.2f}%")
    print(f"  Calmar Ratio : {calmar:.3f}")
    print(f"  Hit Rate     : {hit*100:.1f}%")
    return cum

mom_cum = stats(port_ret, 'Cross-Sectional Momentum (L/S)')
spy_cum = stats(spy_ret,  'SPY Buy & Hold')
```

```python
# --- Plot equity curves ---
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Equity curves
axes[0,0].plot(mom_cum.index, mom_cum.values, label='Momentum L/S', color='darkorange', lw=1.5)
axes[0,0].plot(spy_cum.index, spy_cum.values, label='SPY Buy&Hold',  color='steelblue',  lw=1.2, ls='--')
axes[0,0].set_title('Equity Curves (Growth of $1)', fontsize=12)
axes[0,0].legend()

# Drawdown
mom_dd = (mom_cum - mom_cum.cummax()) / mom_cum.cummax()
axes[0,1].fill_between(mom_dd.index, mom_dd.values * 100, color='red', alpha=0.4)
axes[0,1].set_title('Momentum Strategy Drawdown (%)', fontsize=12)
axes[0,1].set_ylabel('%')

# Annual returns bar chart
ann_rets = port_ret.resample('YE').apply(lambda x: (1+x).prod() - 1)
colors = ['green' if r > 0 else 'red' for r in ann_rets.values]
axes[1,0].bar(ann_rets.index.year, ann_rets.values * 100, color=colors, alpha=0.7)
axes[1,0].axhline(0, color='black', lw=0.8)
axes[1,0].set_title('Annual Returns (%)', fontsize=12)
axes[1,0].set_ylabel('%')

# Rolling 12-month Sharpe
roll_sharpe = (port_ret.rolling(12).mean() * 12) / (port_ret.rolling(12).std() * np.sqrt(12))
axes[1,1].plot(roll_sharpe.index, roll_sharpe.values, color='purple', lw=1)
axes[1,1].axhline(0, color='grey', lw=0.6)
axes[1,1].axhline(1, color='green', ls='--', lw=0.8, label='Sharpe=1')
axes[1,1].set_title('Rolling 12-Month Sharpe', fontsize=12)
axes[1,1].legend()

plt.tight_layout()
plt.savefig('../data/momentum_results.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
# --- Factor decay analysis: how long does momentum persist? ---
hold_periods = range(1, 13)
hold_sharpes = []
for h in hold_periods:
    rets = []
    for i in range(LOOKBACK_MONTHS + 1, len(px_monthly) - h):
        sig_row = stock_signal.iloc[i].dropna()
        if len(sig_row) < 20: continue
        ranked = sig_row.rank(pct=True)
        longs  = ranked[ranked >= 1 - QUINTILE].index
        shorts = ranked[ranked <= QUINTILE].index
        fwd_h  = (px_monthly[stock_cols].iloc[i+h] / px_monthly[stock_cols].iloc[i] - 1)
        r = (fwd_h[longs].mean() - fwd_h[shorts].mean()) / 2
        rets.append(r)
    r_series = pd.Series(rets)
    sh = r_series.mean() / r_series.std() * np.sqrt(12 / h) if r_series.std() != 0 else 0
    hold_sharpes.append(sh)

plt.figure(figsize=(10, 4))
plt.bar(hold_periods, hold_sharpes, color=['green' if s > 0 else 'red' for s in hold_sharpes], alpha=0.7)
plt.axhline(0, color='black', lw=0.8)
plt.xlabel('Holding Period (Months)')
plt.ylabel('Annualised Sharpe')
plt.title('Momentum Factor Decay — Sharpe vs. Holding Period')
plt.tight_layout()
plt.show()
```
